# Datetime EDA

```python
test['date_of_match']= pd.to_datetime(test['date_of_match']) 
```

```python
#extracting Date values to columns month day year

test['month'] = test['date_of_match'].dt.month
test['day'] = test['date_of_match'].dt.day
test['year'] = test['date_of_match'].dt.year
test['quarter'] = test['date_of_match'].dt.quarter
#data['semester'] = np.where(data.quarter.isin([1,2]),1,2)
test['dayofweek'] = test['date_of_match'].dt.dayofweek
test['dayofweek_name'] = test['date_of_match'].dt.day_name()

#(datetime.datetime.today() - data['date_of_match']).head()
```

# "lk1lk2" 1 if difference is positive else 0  

```python
test['lk1lk2'] = test.lk1 - test.lk2


test['lk1lk2'] = np.where(test['lk1-lk2'] < 0, 1, 0).tolist()
```

# lk1-lk2	

```python
test['lk1-lk2'] = test.lk1 - test.lk2
```

```python
test['lk1-lk2'] = np.where(test['lk1-lk2'] == 0.0, 1, test['lk1-lk2']).tolist()
```

# "time_diff"  put 1 if the match happened in the same day

```python
test['time_diff'] = test.day
a = np.array(test['time_diff'].values.tolist())
test['time_diff'] = np.where(test['date_of_match'].dt.day.diff().eq(0) == True, 1, 0).tolist()
```

# "lk2_freq" and "lk1_freq"

```python
color_encode = (test.groupby('lk1').size()) / len(test)

test['lk1_freq'] = test['lk1'].apply(lambda x : color_encode[x])


color_encode = (test.groupby('lk2').size()) / len(test)

test['lk2_freq'] = test['lk2'].apply(lambda x : color_encode[x])
```

# weekend if 5 or 6 put 1 else 0

```python
test['weekend'] = np.where(test['dayofweek_name'].isin(['Sunday','Saturday']),1,0)
```

# "rate_error"

```python
# raplace negative values with 0 and positive with 1
data['rate_error'] = data.day
a = np.array(data['rate_error'].values.tolist())
data['rate_error'] = np.where(((data['match_outcome'] == data['lk1lk2']) == False) == True, 1, 0).tolist()
```
